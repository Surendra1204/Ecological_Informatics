```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load data, skipping the first 6 rows of metadata
file_path = 'pca_fused.asc'
data_matrix = np.loadtxt(file_path, skiprows=6)

# Reshape data into a 2D array for clustering
data_reshaped = data_matrix.flatten().reshape(-1, 1)

# Define the range for clusters
cluster_range = range(1, 11)
inertia_values = []

# Run KMeans clustering for each k and store the inertia
for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(data_reshaped)
    inertia_values.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(10, 6))
plt.plot(cluster_range, inertia_values, marker='o', linestyle='-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid(True)
plt.show()
```
